```python
import pandas as pd
import numpy as np
import time
```

```python
df_stop_101 = pd.read_csv('stopDataByStation/station101Mar7toMar15.csv')
df_stop_102 = pd.read_csv('stopDataByStation/station102Mar7toMar15.csv')
df_stop_103 = pd.read_csv('stopDataByStation/station103Mar7toMar15.csv')
df_sensor = pd.read_csv('sensorDataMar8toMar20.csv')
```

```python
df_stop_101['Starttid'] = df_stop_101['Starttid'].astype('datetime64')
df_stop_101['Sluttid'] = df_stop_101['Sluttid'].astype('datetime64')
df_stop_102['Starttid'] = df_stop_102['Starttid'].astype('datetime64')
df_stop_102['Sluttid'] = df_stop_102['Sluttid'].astype('datetime64')
df_stop_103['Starttid'] = df_stop_103['Starttid'].astype('datetime64')
df_stop_103['Sluttid'] = df_stop_103['Sluttid'].astype('datetime64')
df_sensor['time'] = df_sensor['time'].astype('datetime64')
```

```python
df_sensor.head()
```

```python
df_stop_101 = df_stop_101.drop_duplicates().sort_values(by='Starttid').reset_index().drop(['index','Unnamed: 0'], axis=1)
df_stop_102 = df_stop_102.drop_duplicates().sort_values(by='Starttid').reset_index().drop(['index','Unnamed: 0'], axis=1)
df_stop_103 = df_stop_103.drop_duplicates().sort_values(by='Starttid').reset_index().drop(['index','Unnamed: 0'], axis=1)
df_sensor = df_sensor.sort_values(by='time').reset_index().drop(['index','Unnamed: 0'], axis=1).drop_duplicates().reset_index().drop(['index'], axis=1)
```

```python
df_sensor.head()
```

```python
df_sensor['next_time'] =  df_sensor['time'].shift(-1)
```

```python
df_sensor.head()
```

```python
df_sensor_101 = df_sensor.copy()
df_sensor_102 = df_sensor.copy()
df_sensor_103 = df_sensor.copy()
```

```python
for i, row1 in df_sensor_101.iterrows():
    for j, row2 in df_stop_101.iterrows():
        if row1['time'] <= row2['Starttid'] and row2['Starttid'] <= row1['next_time']:
            df_sensor_101.at[i, 'next_stop_101_time'] = row2['Starttid']
            df_sensor_101.at[i, 'next_stop_101_type'] = row2['Stopporsak']
            df_sensor_101.at[i, 'next_stop_101_information'] = row2['Station']
            
            
for i, row1 in df_sensor_102.iterrows():
    for j, row2 in df_stop_102.iterrows():
        if row1['time'] <= row2['Starttid'] and row2['Starttid'] <= row1['next_time']:
            df_sensor_102.at[i, 'next_stop_102_time'] = row2['Starttid']
            df_sensor_102.at[i, 'next_stop_102_type'] = row2['Stopporsak']
            df_sensor_102.at[i, 'next_stop_102_information'] = row2['Station']
            
for i, row1 in df_sensor_103.iterrows():
    for j, row2 in df_stop_103.iterrows():
        if row1['time'] <= row2['Starttid'] and row2['Starttid'] <= row1['next_time']:
            df_sensor_103.at[i, 'next_stop_103_time'] = row2['Starttid']
            df_sensor_103.at[i, 'next_stop_103_type'] = row2['Stopporsak']
            df_sensor_103.at[i, 'next_stop_103_information'] = row2['Station']
```

```python
df_sensor_103.head()
```

```python
columns_to_fill = ['next_stop_101_time', 'next_stop_101_type', 'next_stop_101_information']
filled_cols = df_sensor_101[columns_to_fill].fillna(method='bfill')
df_sensor_101[columns_to_fill] = filled_cols
df_sensor_101.head()

columns_to_fill = ['next_stop_102_time', 'next_stop_102_type', 'next_stop_102_information']
filled_cols = df_sensor_102[columns_to_fill].fillna(method='bfill')
df_sensor_102[columns_to_fill] = filled_cols

columns_to_fill = ['next_stop_103_time', 'next_stop_103_type', 'next_stop_103_information']
filled_cols = df_sensor_103[columns_to_fill].fillna(method='bfill')
df_sensor_103[columns_to_fill] = filled_cols
df_sensor_103.head()
```

```python
df_sensor_101['RUL_101'] = df_sensor_101['next_stop_101_time'] - df_sensor_101['time']
df_sensor_102['RUL_102'] = df_sensor_102['next_stop_102_time'] - df_sensor_102['time']
df_sensor_103['RUL_101'] = df_sensor_103['next_stop_103_time'] - df_sensor_103['time']
df_sensor_103.head()
```

```python
df_sensor_101.to_csv('df_sensor_stop101')
df_sensor_102.to_csv('df_sensor_stop102')
df_sensor_103.to_csv('df_sensor_stop103')
```
