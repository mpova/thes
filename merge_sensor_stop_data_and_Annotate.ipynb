```python
# Create a new column indicating whether an error occurred at each timestamp
errors = df_stop_101.Stopporsak.unique()
def error_indicator(row):
    if row['Stopporsak'] in errors:
        return 1
    else:
        return 0

df_sensor_stop101['error_indicator'] = df_sensor_stop101.apply(error_indicator, axis=1)
df_sensor_stop102['error_indicator'] = df_sensor_stop102.apply(error_indicator, axis=1)
df_sensor_stop103['error_indicator'] = df_sensor_stop103.apply(error_indicator, axis=1)

# Calculate the remaining useful life for each timestamp
def calculate_RUL(df):
    RUL = np.zeros(len(df))
    last_error_index = -1
    for i, row in df.iterrows():
        if row['error_indicator'] == 1:
            last_error_index = i
        if last_error_index == -1:
            RUL[i] = np.nan
        else:
            RUL[i] = (row['time'] - df.loc[last_error_index]['time'] ).total_seconds()
    return RUL

df_sensor_stop101['RUL'] = calculate_RUL(df_sensor_stop101)
df_sensor_stop102['RUL'] = calculate_RUL(df_sensor_stop102)
df_sensor_stop103['RUL'] = calculate_RUL(df_sensor_stop103)
```

```python
# Define the list of strings of interest
string_list = df_sensor_stop101.Stopporsak.unique().tolist()
string_list = [x for x in string_list if x != 'nan']

# Convert 'Stopporsak' column to string data type and replace NaN values with empty string
df_sensor_stop101['Stopporsak'] = df_sensor_stop101['Stopporsak'].apply(lambda x: str(x)).fillna('')

# Create a new column to store the RUL values
df_sensor_stop101['RUL'] = 0
# Loop through each row in the dataframe
for i, row in df_sensor_stop101.iterrows():
    
    # Check if any of the strings are present in the current row
    if any(s in row['Stopporsak'] for s in string_list):
        df_sensor_stop101.at[i, 'RUL'] = 0  # Set RUL to 0
        
    else:
        next_rows = df_sensor_stop101.loc[i+1:, 'Stopporsak'].apply(lambda x: any(s in x for s in string_list))

        if any(next_rows):
            next_row = next_rows.values.argmax() + i + 1
            time_diff = df_sensor_stop101.loc[next_row, 'time'] - row['time']
            df_sensor_stop101.at[i, 'RUL'] = time_diff
        else:
            df_sensor_stop101.at[i, 'RUL'] = 0
```
